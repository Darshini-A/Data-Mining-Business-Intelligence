```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("/ToyotaCorolla.csv")
data.head()
```

```python
plt.scatter(data['Age_08_04'],data['Price'])
plt.title('Scatter Plot')
plt.xlabel('Age')
plt.ylabel('Price')
plt.colorbar()
plt.show()
```

```python
plt.plot(data['Age_08_04'])
plt.plot(data['Price'])
plt.title('Line Chart')
plt.xlabel('Age')
plt.ylabel('Price')
plt.show()
```

```python
plt.bar(data['Age_08_04'],data['Price'])
plt.title('Bar chart')
plt.xlabel('Age')
plt.ylabel('Price')
plt.show()
```

```python
plt.hist(data['Fuel_Type'])
plt.title("Histogram")
plt.show()
```

```python
sns.barplot(x='Age_08_04',y='Price',data=data)
plt.show()
```

```python
sns.lineplot(x='Age_08_04',y='Price',data=data)
plt.show()
```
